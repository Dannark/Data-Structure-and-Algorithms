```
# O(n) Solution
def uni_char(s):

  map = {}

  for n in s:
    if n not in map:
      map[n] = 1
    else:
      return False
  
  return True

uni_char('abcde')
```

```
# another Solution using Sets
def uni_char(s):
  return len(set(s)) == len(s)
uni_char('abcdee')  
```
