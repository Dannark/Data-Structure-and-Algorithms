```
class Queue(object):
  def __init__(self):
    self.items = []

  def add(self, item):
    self.items.insert(0, item)

  def isEmpty(self):
    return self.items == []

  def pop(self):
    return self.items.pop()

  def size(self):
    return len(self.items)
s = Queue()
s.add('one')
s.add('two')
```
