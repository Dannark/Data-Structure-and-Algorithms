```
# Check if strings are anagrams (My version)
# This is a O(n) algorithim

def isAnagram(s1, s2):
  letters1 = {}
  letters2 = {}

  s1 = s1.lower().replace(' ','')
  s2 = s2.lower().replace(' ','')
  
  if len(s1) != len(s2):
    return False

  for c in s1:
    if not c in letters1:
      letters1[c] = 0
    letters1[c] += 1

  for c in s2:
    if not c in letters2:
      letters2[c] = 0
    letters2[c] += 1
  
  return letters1 == letters2

s1 = 'Banana'
s2 = s1[::-1]

Solution().isAnagram('abc d', 'cdba')
```
